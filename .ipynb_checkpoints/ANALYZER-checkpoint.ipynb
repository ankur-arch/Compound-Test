```python
def returnCsvdir(root, path):
    return "{}{}".format(root,path)    
```

```python
resnet_101_dataframe =  pd.read_csv(returnCsvdir(root, resnet_101_pth))
resnet_50_dataframe =  pd.read_csv(returnCsvdir(root, resnet_50_pth))
resnext_50_dataframe =  pd.read_csv(returnCsvdir(root, resnext_50_pth))
vgg_16_dataframe =  pd.read_csv(returnCsvdir(root, vgg_16_pth))
wide_resnet_dataframe = pd.read_csv(returnCsvdir(root, wide_resnet_pth))
```

```python
def missclassificationFinder(result, name):
    result.reset_index(drop=True, inplace=True)
    result = result.iloc[:,1:].to_numpy()
    np.fill_diagonal(result, 0)
    maxValues = np.amax(result, axis=1)
    maxIndices = np.argmax(result, axis=1)
    print("{} has {} classes".format(name,len(maxValues)))
    # maxindices are misclassified labels
    return maxValues,maxIndices
```

```python
def top_n_mispredicted_classes(values):
    t_values = torch.tensor(values)
    return torch.topk(t_values,10,largest=True,)
```

```python
r101_v, r101_i =  missclassificationFinder(resnet_101_dataframe, name="Resnet 101")
```

```python
r50_v, r50_i =  missclassificationFinder(resnet_50_dataframe, name="Resnet 50")
```

```python
rx_50_v, rx_50_i =  missclassificationFinder(resnext_50_dataframe, name="Resnext 50")
```

```python
vgg16_v, vgg16_i =  missclassificationFinder(vgg_16_dataframe, name="VGG 16")
```

```python
wide_v, wide_i =  missclassificationFinder(wide_resnet_dataframe, name="Wide resnet")
```

```python
def returnMinMisclassifications(index,log=False):
    #0 is resnext
    #1 is vgg16
    #2 is resnet 50
    #3 is resnet 101
    #4 is wide resnet
    LABELS = ["resnext",
    "vgg16",
    "resnet 50",
    "resnet 101",
     "wide resnet"]
    result = [rx_50_v[index],vgg16_v[index],r50_v[index],r101_v[index],wide_v[index]]
    result = np.array(result)
    mini = np.argmin(result)
    if log: 
        print(LABELS[mini]," : ", result[mini])
        #print(LABELS)
        #print(result)
```

```python
def pred_vs_truth(pair,indices):
    truth = pair.indices.numpy()
    predictions = indices[truth]
    return predictions, truth
```

#### Definition of returns

From **top_n_mispredicted_classes**
- indices are the true classes
- values are the misclassification numbers

```python
def pair_image_returner(truth, pred, mapper):
    results = []
    for y_truth,y_pred in zip(truth, pred):
        #print("predicted_image", mapper.iloc[y_pred].labels , "actual_image", mapper.iloc[y_truth].labels)
        #print("predicted", y_pred, "truth", y_truth)
        results.append([(mapper.iloc[y_pred].labels, mapper.iloc[y_pred].images), (mapper.iloc[y_truth].labels, mapper.iloc[y_truth].images) ])
    return results
```

```python
from gradcam.utils import visualize_cam
from gradcam import GradCAM, GradCAMpp
```

```python
#DATA_NORMALIZER(p_result[1])
```

```python
def gradcammer(model,layer,image):
    gradcam = GradCAM(model,layer)
    gradcam_pp = GradCAMpp(model,layer)
    n_image = DATA_NORMALIZER(image)
    expanded_image = n_image.view(1,3,224,224)
    ## gradcam 
    mask_g, _ = gradcam(expanded_image)
    heatmap_g, result_g = visualize_cam(mask_g, expanded_image)
    
    ## gradcam plus plus
    mask_g_p , _ = gradcam_pp(expanded_image)
    heatmap_g_p, result_g_p = visualize_cam(mask_g_p, expanded_image)
    
    return result_g.permute(1,2,0).numpy(), result_g_p.permute(1,2,0).numpy()
```

```python
from tqdm import tqdm
```

```python
def display(model,layer,truth,pred, mapped, title, misclassifications):
    org = pair_image_returner(truth,pred, mapped)
    rows = len(org)
    columns = 4
    labels = []
    results = []
    figure,axe = plt.subplots(rows,columns,figsize=(25,25))
    plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1)
    figure.subplots_adjust(hspace=.5)
    figure.suptitle(title, fontsize=14, fontweight='bold')
                 
    for i,items in enumerate(org,0):
        p_label, p_result = items[0]
        r_label , r_result = items[1]
        labels.append((p_label, r_label))
        results.append((p_result,r_result))
        
    count = 1 
    for i in tqdm(range(len(org))):
        p_l, y_l = labels[i]
        y_p,y = results[i]
        
        
        returnMinMisclassifications(y_l,log=True) 
        
        axe[i][0].imshow(y)
        axe[i][0].set_xlabel('Class {}'.format(y_l))
        axe[i][0].set_ylabel('Class {}'.format(y_l))
        
        
        
        
        axe[i][1].imshow(y_p)
        axe[i][1].set_xlabel('Predicted Class {} | missclassified {} times'.format(p_l,misclassifications[y_l]))
        
        
        
        g, g_pp = gradcammer(model,layer,y_p)
        
        axe[i][2].imshow(g)
        axe[i][2].set_xlabel("{} Gradcam".format(title))
        
        
        axe[i][3].imshow(g_pp)
        axe[i][3].set_xlabel("{} Gradcam++".format(title))
    figure.savefig('{} analysis image '.format(title))
```
